# Project Name: 2 Step Consecutive Recommender

### INTRODUCTION

In this project, I have implemented a 2 step recommender by using Foursquare API data. When the user specifies 2 activities by using 2 keywords (e.g., ('pizza'-'museum'), ('coffee'-'pub')), the script will responde by showing the best possible consecutive scenarios on the map. To many times, I have found myself in the situations where I say "OK, I want to go to that place, but first let's eat/drink something because I am hungry/sleepy.". As far as I know, there is no such app that makes consecutive recommendations. The recommender system also allows user to filter the recommendations further by accepting more entries such as an upper limit for distance and the number of results to be displayed. By adding more features and by working more on it, I believe that a consecutive recommender system can be used by many people.

### DATA and METHODOLOGY

Only the Foursquare API data have been used in this project. 2 types of calls have been made to get the needed information, where first type was the "venues/explore" calls. By this calls, we were able to search around a desired location by refining the results with desired keywords. By this type a call, we ended up with the names, addresses, IDs, and coordinates of the candidate places to be recommended.

After getting the IDs of the candidate places to be recommended, we have requested for the details of these places by "venues/VENUE_ID" calls. From the response of these calls, we have obtained some desired details of venues such as the ratings of the given place and the number of users who rated it so far. Then, by further modifications and attaching these 2 dataframes together properly, we end up with venue names with required information. Then, the results are sorted by their ratings and clipped so that only top rated places are reserved for the user. The resulting dataFrame is shown below.



By implementing the above procedure again for the consecutive step, we get the data that shows step 2 recommendations. Then, these places are shown on map with an easily interpretable fashion.

### RESULTS

When the search is made withing 500 m and 300 m radius, with the keywords ('taco','museum'), the following folium map is acquired. The blue triangle on the center shows the current location of the user. The respectively big circles shows the first step recommendations. As you can see, there are 3 big circles with colors purple, red, and green. These are the 3 highest rated places that are within 500 m to the user. Similarly, the small circles shows the second step recommendations. These locations are within 300 m to their corresponding first step locations. For instance, 3 small purple circles are the museums that user can go after eating taco in the big purple circle. The details (such as address and rating) of all venues are also shown in popups after clicking on the circles. 



### DISCUSSION

The script can be improved a lot further. For instance, most popular tips about the resulting venues can be added to the popups. In fact, I acquired added the most popular 2 tips for each venue on the dataFrame, but the popup window did not allow me to add such a long texts in it. Therefore, I had to remove them later. There are some other know techniques to expand the popup further, but I didn't want to deal with it all.

Furthermore, information about walking distance (not air distance) may be a good detail to add into the popups. This information may be obtained by using other types of Map APIs.

### CONCLUSION

In this project, a 2 step recommender system is built. By taking the maximum distances, keywords, and number of places as input, the system makes recommendations on the map. Although there are still a lot to improve, the scprit seems to work well.
